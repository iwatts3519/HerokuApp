```python
import mysql.connector
import pandas as pd

from wordcloud import WordCloud,STOPWORDS
import matplotlib.pyplot as pyplot



mydb = mysql.connector.connect(
  host="localhost",
  user="root",
  passwd="",
  database="bookmein"
)
```

```python
data = pd.read_sql_query('SELECT DISTINCT exhibitor_message.message AS messages FROM exhibitor_message', mydb)

data['messages'] = data['messages'].str.replace('\n','')
data['messages'] = data['messages'].str.replace('\r','')


my_stop_words = ["Regards", "Hi", "Hello", "Kind Regards",
                 "Best Regards", "helloo", "Thanks", "Kind",
                 "well","best","test","testTest", "morning","afternoon"] + list(STOPWORDS)


text_data = data.values
wordcloud = WordCloud(
    height = 500,
     width = 750,
    background_color = '#000000',
    stopwords = my_stop_words).generate(str(text_data))
figure = pyplot.figure(
    figsize = (10,7.5),
    facecolor = '#000000',
    edgecolor = '#000000')
pyplot.imshow(wordcloud, interpolation = 'bilinear')
pyplot.axis('off')
pyplot.tight_layout(pad=0)
pyplot.savefig("assets/wordcloud1.png")

data
```
